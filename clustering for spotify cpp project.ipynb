```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/svyat/Downloads/playlist_2010to20222.csv', sep=';')
df
```

```python
df.artist_genres.apply(lambda x: [eval(x)])
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
genres_df = pd.DataFrame(mlb.fit_transform(df.artist_genres.apply(lambda x: eval(x))), columns=mlb.classes_)
```

```python
mlb.classes_
```

```python
# new_df = pd.concat([df.track_name, df.iloc[:, 10:-2]],axis=1)
new_df = df.iloc[:, 10:-2].fillna(0)
new_df = pd.concat([new_df, genres_df], axis=1)
# new_df = genres_df
new_df
```

```python
# kmeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=50,random_state=0).fit(new_df)
df["clusters"] = kmeans.labels_
df.sort_values("clusters").head()
```

```python
# spectral
from sklearn.cluster import SpectralClustering
clustering = SpectralClustering(n_clusters=50,
        assign_labels='discretize',
        random_state=2024).fit(new_df)
clustering.labels_
df["s_clusters"] = clustering.labels_
df.sort_values("s_clusters").head()
```

```python
df.to_csv("clusters_50.csv", sep=';')
```
